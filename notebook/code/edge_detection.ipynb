```python
img_BGR = cv2.imread("./dataset/working/preparing/probe.jpg")
img_gray = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2GRAY)

# cv2.imshow('Original', img_BGR)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
cv2.imwrite("./temp/01-original.jpg", img_BGR)
plt.imshow(img_BGR)
plt.title('Original')
plt.show()

cv2.imwrite("./temp/02-gray.jpg", img_gray)
plt.imshow(img_gray)
plt.title('Gray')
plt.show()
```

```python
print("Canny Edge Detection")

for (img, name) in imgs:
    edges = cv2.Canny(image=img, threshold1=100, threshold2=200) # Canny Edge Detection
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))
    fig.suptitle(name)
    ax1.set_title("Original")
    ax1.imshow(img)
    ax2.set_title("Canny")
    ax2.imshow(edges)
```
