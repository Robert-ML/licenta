```python
import cv2
import numpy as np
import typing
from matplotlib import pyplot as plt
```

```python
img_BGR = cv2.imread("/mnt/HD/School/A4/licenta/notebook/code/dataset/masini/000/20211021_130952.jpg")
img_gray = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2GRAY)

# cv2.imshow('Original', img_BGR)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,7))
fig.suptitle('Original')
ax1.imshow(img_BGR)
ax1.set_title('BGR')
ax2.imshow(cv2.cvtColor(img_BGR, cv2.COLOR_BGR2RGB))
ax2.set_title('RGB')
```

```python

# They say soble works better on burred images
img_blur = cv2.GaussianBlur(img_gray, (13,13), 0)

imgs = [(img_gray, "gray"), (img_blur, "blur")]

for (img, name) in imgs:
    fig, ax = plt.subplots(1, 1, figsize=(8,7))
    ax.set_title(name)
    ax.imshow(img)
```

```python
print("Sobel Edge Detection")
ksize = 5
for (img, name) in imgs:
    sobelx = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=ksize) # Sobel Edge Detection on the X axis
    sobely = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=ksize) # Sobel Edge Detection on the Y axis
    sobelxy = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=ksize) # Combined X and Y Sobel Edge Detection

    fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(23,7))
    fig.suptitle(name)
    ax1.set_title("Original")
    ax1.imshow(img)
    ax2.set_title("Sobel X")
    ax2.imshow(sobelx)
    ax3.set_title("Sobel Y")
    ax3.imshow(sobely)
    ax4.set_title("Combined")
    ax4.imshow(sobelxy)
    cv2.imwrite("./" + name + ".jpg", sobelxy)
```

```python
print("Canny Edge Detection")

for (img, name) in imgs:
    edges = cv2.Canny(image=img, threshold1=100, threshold2=200) # Canny Edge Detection
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))
    fig.suptitle(name)
    ax1.set_title("Original")
    ax1.imshow(img)
    ax2.set_title("Canny")
    ax2.imshow(edges)
```
